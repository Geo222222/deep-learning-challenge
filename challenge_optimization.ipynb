```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
application_df = application_df.drop(columns = ['EIN'])
application_df.head()
```

```python
for x in application_df.columns:
    print(x, len(application_df[x].unique()))
```

```python
# Look at NAME counts for the binning process
counts1 = application_df['NAME'].value_counts()
counts1
```

```python
# use the variable name names_replace
names_replace = list(counts1[counts1<5].index)
```

```python
# Replace in dataframe
for name in names_replace:
    application_df['NAME'] = application_df['NAME'].replace(name,"Other")

# Check to make sure binning was successful
application_df['NAME'].value_counts()
```

```python
#classification counts
countsbinning = application_df['CLASSIFICATION'].value_counts()
countsclassification = countsbinning[countsbinning>1]
countsclassification
```

```python
# use the variable name `classifications_to_replace` with less than 100
classifications_to_replace = list (countsbinning[countsbinning<100].index)
classifications_to_replace
```

```python
# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df = pd.get_dummies(application_df,dtype=float)
application_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = application_df['IS_SUCCESSFUL'].values
y
X = application_df.drop('IS_SUCCESSFUL', axis=1).values
X
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state = 42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization

number_input_features = len( X_train_scaled[0])

hidden_nodes_layer1=3
hidden_nodes_layer2=14
hidden_nodes_layer3=21

nn = tf.keras.models.Sequential()
nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation='relu'))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

nn.add(Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
nn.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics=['accuracy'])
# Train the model

fit_model = nn.fit(X_train_scaled,y_train,validation_split=0.15, epochs=100)
```

```python
# Export our model to HDF5 file
model_loss, model_accuracy = nn.evaluate(X_test_scaled, y_test, verbose=2)
# Save the model to an H5 file
nn.save("mod_21_opti.h5")
```
